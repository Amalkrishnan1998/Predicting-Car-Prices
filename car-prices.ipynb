**Critical Thinking Questions:**

1.Why is handling missing data important?

Handling missing data is essential because missing values can lead to inaccurate statistical analysis, biased conclusions, and poor machine learning performance. Proper techniques such as mean replacement for numerical features and mode replacement for categorical features help preserve data quality, improve model reliability, and ensure consistent analysis results.

2.Which method is better: dropping or replacing? Why? 

Replacing missing values is generally preferred over dropping records because it preserves more data and reduces information loss. Mean replacement is suitable for numerical variables, while mode replacement is appropriate for categorical variables. However, when the missing values occur in the target variable (such as price), dropping those records is often the better choice because the missing target cannot be reliably estimated. Therefore, a combination of replacement and selective dropping is usually the most effective approach.

3.How does binning help analysis? 

Binning helps analysis by converting continuous numerical data into meaningful categories. It simplifies interpretation, reduces noise, improves visualizations, and makes patterns easier to identify. In the automobile dataset, horsepower was grouped into Low, Medium, and High categories, enabling easier comparison of vehicle characteristics and prices across different performance levels.

Step 2:Add more features later for better performance.

```python
X = df[['horsepower', 'engine-size']]
y = df['price-category']
```

Step 3: Train-Test Split (80/20)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

Step 4: Train Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)
```

Step 5: Make Predictions

```python
y_pred = log_model.predict(X_test)
```

Step 6: Evaluate Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", round(accuracy, 4))
```

Step 7: Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

Step 8: Visualize Confusion Matrix

```python
plt.figure(figsize=(6,4))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Low','Medium','High'],
    yticklabels=['Low','Medium','High']
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

Model performance was evaluated using Accuracy and a Confusion Matrix. The results indicate the model's ability to correctly classify vehicles into price categories based on their technical characteristics.
